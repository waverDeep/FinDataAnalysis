```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.family"] = 'NanumBarunGothic'

sigoongu = list(set(df_1903['고객소재지_시군구']))
print(len(sigoongu),'개')
print(sigoongu)

large_category = list(set(df_1903['품목대분류명']))
print(len(large_category),'개')
print(large_category)


plt.figure(figsize=(100,70))
for index, gu in enumerate(sigoongu):
    base = [ 0 for i in range(len(large_category))]
    temp = df_1903[df_1903['고객소재지_시군구']==gu]
    temp = temp['품목대분류명'].value_counts()
    for i, t in enumerate(temp):
        base[i] += t
        plt.subplot(5,5,index+1)
        plt.bar(large_category, base, color='blue', alpha=0.5)
        plt.xticks(rotation=90)
        plt.title(gu)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.family"] = 'NanumBarunGothic'

sigoongu = list(set(df_1903['고객소재지_시군구']))
print(len(sigoongu),'개')
print(sigoongu)

middle_category = list(set(df_1903['품목중분류명']))
print(len(middle_category),'개')
print(middle_category)


plt.figure(figsize=(100,70))
for index, gu in enumerate(sigoongu):
    base = [ 0 for i in range(len(middle_category))]
    temp = df_1903[df_1903['고객소재지_시군구']==gu]
    temp = temp['품목중분류명'].value_counts()
    for i, t in enumerate(temp):
        base[i] += t
        plt.subplot(5,5,index+1)
        plt.bar(middle_category, base, color='blue', alpha=0.5)
        plt.xticks(rotation=90)
        plt.title(gu)
plt.show()
```

```python
df_1903_매출금액합 = df_1903.groupby('고객소재지_시군구')['매출금액'].sum()
print(df_1903_매출금액합)
df_1903_매출금액합.plot.bar()
```
